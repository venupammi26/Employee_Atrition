```python
new_data=data1.drop(['JobLevel','EmployeeCount','EmployeeNumber'],axis=1)
new_data.head()
```

# PCA

```python
X= new_data.iloc[:,:-1].values
y=new_data.iloc[:,-1].values
```

```python
#Feature Scaling

from sklearn.preprocessing import StandardScaler

SS= StandardScaler()
scaled_data=SS.fit_transform(X)

scaled_data=pd.DataFrame(scaled_data,columns=new_data.iloc[:,:-1].columns)
scaled_data.head()
```

```python
from sklearn.decomposition import PCA 
  
# Let's say, components = 5
pca = PCA(n_components = 5) 
pca.fit(scaled_data) 
x_pca = pca.transform(scaled_data) 
  
x_pca.shape 
```

```python
#Data Splitting into train and test
X=x_pca

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=1)
```

# Logistic Regrssion

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

LR=LogisticRegression()
LR.fit(X_train,y_train)

y_pred=LR.predict(X_test)
accuracy_score(y_pred,y_test)
```

# Support Vector Machine

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

SV=SVC(kernel='linear')
SV.fit(X_train,y_train)

y_pred=SV.predict(X_test)
accuracy_score(y_pred,y_test)
```

```python
SV=SVC(kernel='poly', degree=3)
SV.fit(X_train,y_train)

y_pred=SV.predict(X_test)
accuracy_score(y_pred,y_test)
```

```python
SV=SVC(kernel='rbf')
SV.fit(X_train,y_train)

y_pred=SV.predict(X_test)
accuracy_score(y_pred,y_test)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
KNN=KNeighborsClassifier(n_neighbors=9)
KNN.fit(X_train,y_train)

y_pred=KNN.predict(X_test)
accuracy_score(y_pred,y_test)
```

```python
i=3

while i<=21:
    KNN=KNeighborsClassifier(n_neighbors=i)
    KNN.fit(X_train,y_train)
    y_pred=KNN.predict(X_test)
    print('accuracy score with neighbours k=',i,':  ',accuracy_score(y_pred,y_test))
    i+=2

```

# Naive bayes

```python
from sklearn.naive_bayes import GaussianNB
gnn=GaussianNB()

gnn.fit(X_train,y_train)

y_pred=gnn.predict(X_test)
accuracy_score(y_pred,y_test)
```

# Desision Tree

```python
from sklearn.tree import DecisionTreeClassifier
DTC=DecisionTreeClassifier()
DTC.fit(X_train,y_train)

y_pred=DTC.predict(X_test)
accuracy_score(y_pred,y_test)
```

```python
#Heprt parameter tuning

from sklearn.model_selection import RandomizedSearchCV 

param_dist = {"max_depth": [3, None], 
              "criterion": ["gini", "entropy"]} 
  
# Instantiating Decision Tree classifier 
DTC = DecisionTreeClassifier() 
  
# Instantiating RandomizedSearchCV object 
DTC = RandomizedSearchCV(DTC, param_dist, cv = 5) 
  
DTC.fit(X_train,y_train)
y_pred=DTC.predict(X_test)
accuracy_score(y_pred,y_test)  
```

# Random Forest


```python
from sklearn.ensemble import RandomForestClassifier

RFC=RandomForestClassifier(n_estimators=50)

RFC.fit(X_train,y_train)

y_pred=RFC.predict(X_test)
accuracy_score(y_pred,y_test)
```

```python
#Hyper parameter tuning
grid_param ={
              'n_estimators': [50, 75, 100, 125, 150, 175],
              'min_samples_split':[2,4,6,8,10],
              'min_samples_leaf': [1, 2, 3, 4],
              'max_depth': [5, 10, 15, 20, 25],
              'criterion': ['gini', 'entropy'],
              'bootstrap': [True, False]
            }

from sklearn.model_selection import GridSearchCV

gd_sr = GridSearchCV(estimator=RFC,
                     param_grid=grid_param,
                     scoring='accuracy',
                     cv=5,
                     n_jobs=-1)

gd_sr.fit(X_train, y_train)

best_parameters = gd_sr.best_params_
print(best_parameters)

best_result = gd_sr.best_score_
print(best_result)
```

```python
#Final model

RFC=RandomForestClassifier(n_estimators=100, bootstrap=True, criterion='entropy')

RFC.fit(X_train,y_train)

y_pred=RFC.predict(X_test)
accuracy_score(y_pred,y_test)
```
